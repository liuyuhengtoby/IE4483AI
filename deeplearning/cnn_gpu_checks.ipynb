# GPU Check (only applicable to my own local setup)

This notebook checks the current environment and gpu allocation to ensure the correct setup for future training

```python
import tensorflow as tf, os
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"  # reduce log noise

gpus = tf.config.list_physical_devices('GPU')
if gpus:
    for gpu in gpus:
        tf.config.experimental.set_memory_growth(gpu, True)
    print(f"✅ Using GPU: {gpus[0].name}")
else:
    print("⚠️ No GPU detected. Running on CPU.")

print("TensorFlow:", tf.__version__, " | CUDA runtime:", tf.sysconfig.get_build_info()['cuda_version'])
```
