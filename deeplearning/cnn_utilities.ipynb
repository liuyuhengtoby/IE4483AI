# Some Utilies

This notebook stores section of code for utilities such as section data to smaller set

```python
import os, random, shutil
SRC_TRAIN_DIR = "../datasets/train"
DEST_DIR = "../datasets_small"
CLASSES = ['dog', 'cat']

def make_small_dataset(src_dir, dest_dir, n_train=500, n_val=100, overwrite=True):
    os.makedirs(dest_dir, exist_ok=True)
    for split in ['train', 'val']:
        for cls in CLASSES:
            os.makedirs(os.path.join(dest_dir, split, cls), exist_ok=True)

    for cls in CLASSES:
        src_cls_dir = os.path.join(src_dir, cls)
        files = os.listdir(src_cls_dir)
        random.shuffle(files)
        train_files = files[:n_train]
        val_files = files[n_train:n_train+n_val]

        for split, flist in [('train', train_files), ('val', val_files)]:
            dst_cls_dir = os.path.join(dest_dir, split, cls)
            for f in flist:
                shutil.copy(os.path.join(src_cls_dir, f),
                            os.path.join(dst_cls_dir, f))

make_small_dataset(SRC_TRAIN_DIR, DEST_DIR, n_train=500, n_val=100)
print("âœ… Small dataset created at:", DEST_DIR)
```
